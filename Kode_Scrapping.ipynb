# Data Scrapping

```python
from google_play_scraper import app, reviews_all, Sort

scrapreview, _ = reviews(
    'com.rtsoft.growtopia',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000
)
```

```python
import csv

# Buka file CSV satu kali
with open('review_growtopia_mobile_app.csv', mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Iterasi langsung pada scrapreview (list review)
    for review in scrapreview:
        if isinstance(review, dict):
            content = review.get('content', '')
            writer.writerow([content])
        else:
            print(f"Invalid review format: {review}")

print("CSV file updated successfully!")

import os
print("Current working directory:", os.getcwd())

# Jika kamu menggunakan Google Colab dan ingin mengupload ke Google Drive:
from google.colab import drive
drive.mount('/content/drive')
!cp review_growtopia_mobile_app.csv "/content/drive/MyDrive/Colab Notebooks/Proyek Analisis Sentimen/"

print("Jumlah review yang diambil:", len(scrapreview))
```
